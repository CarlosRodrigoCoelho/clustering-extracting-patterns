# Preprocessing data
    The next step is to normalize our data. In some attributes related to frequency, we will have a limit that varies from 0 to 1, where 0 is 0% frequency and 1, in turn, means 100%. For the balance attribute, we will not have the same minimum or maximum organization, and this can cause the algorithm to generate unsatisfactory data for us.

    To normalize all data between zero and one, we will use the Scikit learn library, importing the Normalizer from the sklearn.preprocessing package.

```python
df1.values
```

```python
from sklearn.preprocessing import Normalizer
values = Normalizer()
values_normalized = values.fit_transform(df1.values)
values_normalized
```

# Clustering

```python
from sklearn.cluster import KMeans

SEED = 123
np.random.seed(SEED)

kmeans = KMeans(n_clusters=5, n_init=10, max_iter=300, random_state=SEED)
y_pred = kmeans.fit_predict(values_normalized)
y_pred
```

# Cluster Validation

#### Silhouette coefficient
    I'm gonna use the Silhouette coefficient, itis a metric for evaluating clustering in machine learning, which measures the quality of clustering in terms of how well points belong to their cluster and how dissimilar they are to other clusters. The value of the Silhouette coefficient ranges from -1 to 1, where values ​​closer to 1 indicate good clustering and values ​​closer to -1 indicate poor clustering.

```python
from sklearn import metrics

# creating labels generated by KMeans
labels = kmeans.labels_

silhouette = metrics.silhouette_score(values_normalized, labels, metric='euclidean')

print(f'The silhouette mean is: {silhouette}')
```

The closer the silhouette value is to 1, the more separated the clusters are.

#### Davies–Bouldin index
    The Davies-Bouldin index is a metric used to evaluate the clustering quality of a dataset. It measures the average similarity between each cluster and its most similar cluster, taking into account the cluster centroids and the distance between them. A lower Davies-Bouldin index (near to 0) indicates better clustering quality, as it means the clusters are well-separated and distinct from each other.

```python
from sklearn import metrics

dbs = metrics.davies_bouldin_score(values_normalized, labels)
print(f'The Davies-Bouldin index is: {dbs}')
```

#### Calinski-Harabasz index
    The Calinski-Harabasz index is a clustering evaluation metric that measures the ratio of between-cluster variance to within-cluster variance. This index is used to evaluate the quality of clustering by measuring the degree of separation between clusters and the compactness of points within each cluster.
    A higher value of CH(K) indicates better separation between clusters and greater compactness within each cluster. However, it should be noted that the choice of the optimal number of clusters is still a challenge, as the CH(K) value may increase or decrease as the number of clusters increases.


```python
from sklearn import metrics

calinski = metrics.calinski_harabasz_score(values_normalized, labels)
print(f'The Calinski-Harabasz index is: {calinski}')
```

# Relative Evaluation
    I have three values ​​to interpret with different metrics, but how do we do that?
    

### Relative Evaluation 1
    We compare the value of metrics based on different configurations of our algorithm. The idea is that in this way we adjust the parameters of the algorithm to obtain the best results.

```python
def clustering_algorithm(n_clusters, dataset):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, max_iter=300,random_state=SEED)
    labels = kmeans.fit_predict(dataset)
    silhouette = metrics.silhouette_score(dataset, labels, metric='euclidean')
    dbs = metrics.davies_bouldin_score(dataset, labels)
    calinski = metrics.calinski_harabasz_score(dataset, labels)
    return silhouette, dbs, calinski
```

```python
silhouette1, dbs1, calinski1 = clustering_algorithm(3, values_normalized)
print(silhouette1, dbs1, calinski1)
```

```python
silhouette2, dbs2, calinski2 = clustering_algorithm(5, values_normalized)
print(silhouette2, dbs2, calinski2)
```

```python
silhouette3, dbs3, calinski3 = clustering_algorithm(10, values_normalized)
print(silhouette3, dbs3, calinski3)
```

```python
silhouette4, dbs4, calinski4 = clustering_algorithm(50, values_normalized)
print(silhouette4, dbs4, calinski4)
```

Based on the results, I will use the 5 cluster configuration, as we get a good result on the silhouette index, which performance is to be evaluated solely on unsupervised clustering performance, so it is interesting to use it as the main gauge.

In general, the choice of which metric to use depends on the specific task and the available information. If the ground truth is known, Calinski-Harabasz may be preferred. If the number of clusters is known, Davies-Bouldin may be preferred. If the algorithm's performance is to be evaluated solely on unsupervised clustering performance, Silhouette may be preferred. Ultimately, it's important to use a combination of metrics and qualitative evaluation to determine the best clustering algorithm for the specific task at hand.

### Relative Evaluation 3 - Cluster Stability
    The idea is that we divide the database and run kmeans for each of these divisions in order to obtain similar results for the metrics, as we want kmeans to behave similarly in each one in order to achieve validation.

```python
set1, set2, set3 = np.array_split(values_normalized, 3)

silhouette1, dbs1, calinski1 =  clustering_algorithm(5, set1)
silhouette2, dbs2, calinski2 =  clustering_algorithm(5, set2)
silhouette3, dbs3, calinski3 =  clustering_algorithm(5, set3)

print(silhouette1, dbs1, calinski1)
print(silhouette2, dbs2, calinski2)
print(silhouette3, dbs3, calinski3)
```


We will have similar values ​​in the three columns and with different metrics, this indicates that our cluster is valid.

### Trying most significant attribute and analyze it

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

As we will analyze the variance, we will look for an attribute in which the values ​​are different between the clusters, as this will increase the chance of observing the particularities of each group.

Those attributes that have the highest variance will be used.

##### Selecting Attributes
    we will collect the amount of attributes in centroids[] with the value 0 that means the first line, and everything will be in max, the maximum value of attributes.
    We'll do a for up to the maximum attribute value range(max) and then we'll print() the name of each of the attributes by writing dataframe.columns.values[i].

    Then we can calculate the variance with .var().

```python
max = len(centroids[0])
for i in range(max):
    print(df1.columns.values[i], '\n{:.4F}'.format(centroids[:, i].var()))
```
